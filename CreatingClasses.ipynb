```python
class Book:
    def __init__(self, title, author, pages):
        self.title = title
        self.author = author
        self.pages = pages

    def description(self):
        return f"{self.title} by {self.author}"

    def is_long(self):
        if self.pages >= 300:
          return "is long"
        else:
          return "is not long"
```

```python
book1 = Book("The Great Gatsby", "F. Scott Fitzgerald", 180)
print(f"{book1.description()} is long? {book1.is_long()}")
```

```python
class Car:
    def __init__(self,brand,model,mileage):
        self.brand = brand
        self.model = model
        self.mileage = mileage
    def drive(self, miles):
        self.mileage += miles
    def info(self):
        return f"{self.brand} {self.model} with {self.mileage} miles."
```

```python
import math
class Circle:
    def __init__(self,radius):
        self.radius = radius
    def area(self):
        return math.pi * (self.radius ** 2)
    def circumference(self):
        return 2 * math.pi * self.radius
```

```python
Circle1 = Circle(4.8)
print(f"Area is {Circle1.area()} and Circumference is {Circle1.circumference()}")
```

```python
class Movie:
    def __init__(self,title,genre,rating):
        self.title = title
        self.genre = genre
        self.rating = rating
    def recommend(self):
        if self.rating >= 8:
          return "Yes"
        else:
          return "No"
    def summary(self):
        return f"{self.title} is a {self.genre} movie with a rating of {self.rating}/10."
```

```python
Movie1 = Movie("Avengers End Game","Action/SciFi",8.4)
print(f"{Movie1.summary()} is recommended? {Movie1.recommend()} ")
```

```python
class Counter:
    def __init__(self):
        self.count = 0
    def increment(self):
        self.count += 1
    def reset(self):
        self.count = 0
    def get_count(self):
        return self.count
```

```python
Counter1 = Counter()
print(f"Initial count: {Counter1.get_count()}")
Counter1.increment()
Counter1.increment()
print(f"Count after two increments: {Counter1.get_count()}")
Counter1.reset()
print(f"Count after reset: {Counter1.get_count()}")
```

```python
class Vehicle:
  def __init__(self, brand, speed):
    self.brand = brand
    self.speed = speed

  def Accelerate(self, speedChange):
    self.speed += speedChange
  def Info(self):
    return f"the vehicle is made by {self.brand} and has a speed {self.speed}"

class Truck(Vehicle):
  def __init__(self, brand, speed, capacity):
     super().__init__(brand, speed)
     self.capacity = capacity

  def load_info(self):
    return self.capacity
```

```python
v1 = Vehicle("Ford", 120)
v1.Info()
```

```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

    def description(self):
        return self.title + " by " + self.author

class Library:
    def __init__(self, name):
        self.name = name
        self.book_list = []

    def add_book(self, book):
        self.book_list.append(book)

    def list_books(self):
        titles = []
        for b in self.book_list:
            titles.append(b.title)
        return titles


lib = Library("Hogwarts Library")
b1 = Book("Spell casting", "Harry Potter")
lib.add_book(b1)
print("Library:", lib.name)
print("Books:", lib.list_books())
```

```python
class Player:
    def __init__(self, name, position):
        self.name = name
        self.position = position

class Team:
    def __init__(self, name):
        self.name = name
        self.players = []

    def add_player(self, p):
        self.players.append(p)

    def team_roster(self):
        roster_names = []
        for p in self.players:
            roster_names.append(p.name)
        return roster_names

my_team = Team("CLE BROWNS")
p1 = Player("Andrew", "WR")
my_team.add_player(p1)
print("Team:", my_team.name)
print("Players:", my_team.team_roster())
```

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return "Small sound"

    def info(self):
        return self.name

class Cat(Animal):
    def __init__(self, name):
        super().__init__(name)

    def speak(self):
        return self.name + " meow purr meow"

    def scratch(self):
        return self.name + " is scratching the carpet"

# testing
c = Cat("Furball")
print("Name:", c.info())
print("Sound:", c.speak())
print("Action:", c.scratch())
```

```python
class Person:
    def __init__(self, name):
        self.name = name

    def introduce(self):
        return "Hi, my name is " + self.name

class Student(Person):
    def __init__(self, name, major):
        super().__init__(name)
        self.major = major

    def introduce(self):
        return "Hi, I'm " + self.name + " and I study " + self.major

class Professor(Person):
    def __init__(self, name, department):
        super().__init__(name)
        self.department = department

    def introduce(self):
        return "I am Professor " + self.name + " from " + self.department

s1 = Student("Andrew Oshatyuk", "ISM")
p1 = Professor("Prof Liu", "Systems Design")

print(s1.introduce())
print(p1.introduce())
```

```python
class account:
  def __init__(self, acctno, balance):
    self.accno = acctno
    self.balance = balance

  def Withdraw(self, amt):
    if amt < self.balance:
      self.balance -= amt
    else:
      return "insufficent amount"

  def deposite(self, amt):
    if amt > 0:
      self.balance += amt
    else:
      return "cannot deposite negative amount"
```
